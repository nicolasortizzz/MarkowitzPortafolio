 Output Gap estimado por OECD:


```python
import pandas as pd
import matplotlib.pyplot as plt

ruta_csv = "Graficos/OECDevolucion.csv"

df_oecd = pd.read_csv(ruta_csv)
if "Date" in df_oecd.columns:
    df_oecd["Date"] = pd.to_datetime(df_oecd["Date"])
else:
    raise ValueError("El archivo no contiene una columna 'Date'.")

if "Value" not in df_oecd.columns:
    raise ValueError("El archivo no contiene una columna 'Value'.")

df_oecd = df_oecd.sort_values("Date").reset_index(drop=True)

plt.figure(figsize=(12,5))
plt.plot(df_oecd["Date"], df_oecd["Value"], marker='o', linewidth=1, markersize=4, label='Value')
plt.title('Evolucin del Output Gap OECD (archivo OECDevolucion)')
plt.xlabel('Fecha')
plt.ylabel('Value')
plt.grid(alpha=0.3)
plt.legend()
ax = plt.gca()
ax.set_xticks(df_oecd["Date"])
ax.set_xticklabels(df_oecd["Date"].dt.strftime('%Y-%m-%d'), rotation=90, ha='center')
plt.tight_layout()
plt.show()
```
